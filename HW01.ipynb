## Задание 1. 
В этом задании вам необходимо составить свой дизайн A/B-теста. Для этого вам нужно выбрать произвольный продукт (например - это маркет-плейс, сервис рекомендации видео, социальная сеть, мобильное приложение или проект, над которым работаете), дать его краткое описание (основная функция продукта и его цель). Далее к этому продукту необходимо сформулировать дизайн A/B-теста, то есть определить следующие пункты:
Изменение - она же гипотеза, которая сделает продукт лучше.
Метрики, по которым будет определяться, что продукт стал лучше.
Аудитория - на ком должна быть протестирована гипотеза.
Инфраструктура - как будут доставляться изменения и как будет деление аудитории происходить. (Описать примерно)
Длительность проведения теста. (Пока не умеем ее рассчитывать, поэтому указать исходя из логических соображение)
Ожидаемый эффект - на сколько должны вырасти ключевые метрики.

### Продукт 
Рассмотрим приложение знакомств (например, Tinder).

Первое впечатление играет порой определяющую роль при знакомстве. Важно начать диалог интересно, чтобы привлечь внимание второй половинки ;-)

### Дизайн А/Б-теста:
1. Гипотеза: добавление приветственных сообщений-подсказок (вместо "привет, как дела?)") повысит вовлеченность пользователей (мужского пола)
2. Метрики: среднее количество сообщений внутри одного диалога (как следствие рост вовлеченности)
3. Аудитория: мужчины, пользующиеся сервисом больше месяца с высокой частотой (эмпирически нужно считать пороговое значение) встречаемости "тривиальных" первых сообщений (в предположении, что у нас есть Берт, который классифицирует первые сообщения на "тривиальность" и у кого высокий процент таких сообщений, тот наша целевая аудитория)
4. Инфраструктура: разделим аудиторию с соблюдением равной пропорции возраста внутри группы (stratified)
5. Длительность: две недели
6. Ожидаемый эффект: +20% к количеству сообщений внутри диалога
